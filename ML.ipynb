```python
import numpy as np
import pandas as pd
from sklearn.svm import SVR
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
import time
from joblib import dump, load
```

```python
# topology features
```

```python
topo_train = pd.read_csv("data/topo_train22.csv")
topo_test_X = pd.read_csv("data/topo_test_X.csv")

topo_train_X = topo_train.iloc[:, :-1]
train_y = topo_train.iloc[:, -1]
```

```python
topo_train
```

```python
scaler = StandardScaler()
scaler.fit(topo_train_X)
scaled_topo_train_X = scaler.transform(topo_train_X)
scaled_topo_test_X = scaler.transform(topo_test_X)
```

```python
# svr 
start = time.time()
svr = SVR(kernel ='rbf',degree = 3,gamma ='auto',coef0 = 0.0,tol = 0.05,C = 1.0,epsilon = 0.1)
svr.fit(scaled_topo_train_X, train_y)
end = time.time()
end-start
```

```python
predictions = svr.predict(scaled_topo_test_X)
predictions
```

```python
result = pd.DataFrame({"Id":range(1,2001), "Predicted": predictions})
result.to_csv("models/result.csv", index=False)
```

```python
dump(svr, "models/model.joblib")
```
